```python

# Extract the @context array from the context_data
combined_context_list = context_data.get("@context", [])
# Create a new document by copying the original dataset_record and
# adding the @context to it.
jsonld_document = dataset_record.copy()

# Add a base IRI to help resolve relative IRIs within the document if any.
# This is good practice for Linked Data.
jsonld_document["@context"] = combined_context_list
```

```python
expanded_data = jsonld.expand(jsonld_document)
```

```python
nquads = jsonld.to_rdf(jsonld_document, {'format': 'application/n-quads'})
print("\n--- RDF N-Quads ---")
print(nquads)
```

```python
from rdflib import Graph # Import Graph from rdflib
# Create an RDFLib Graph
g = Graph()

# Parse the N-Quads into the RDFLib Graph
g.parse(data=nquads, format='nquads')

# Serialize the RDFLib Graph to Turtle format
turtle_data = g.serialize(format='turtle')

print("--- RDF in Turtle Format ---")
print(turtle_data)
print("\n" + "="*50 + "\n")
```

```python
import json
import requests
import yaml
from jsonschema import validate, RefResolver

# Load the GeoJSON data
# with open(r'..\\_sources\\dcat-records\\examples\\bag.geojson', 'r', encoding='utf-8') as f:
with open(r'records-example.json', 'r', encoding='utf-8') as f:
    contents = json.load(f)

# Load the schema
with open('schema.json', 'r', encoding='utf-8') as f:
    schema = json.load(f)

# Function to fetch and parse remote schemas
def fetch_schema(url):
    response = requests.get(url)
    response.raise_for_status()
    if url.endswith('.yaml') or url.endswith('.yml'):
        return yaml.safe_load(response.text)
    else:
        return response.json()

# Create a RefResolver with a handler for HTTP/HTTPS
resolver = RefResolver(base_uri=schema.get('$id', ''), referrer=schema, handlers={'https': fetch_schema, 'http': fetch_schema})

# Validate
try:
    validate(instance=contents, schema=schema, resolver=resolver)
    print("Validation successful: contents conform to schema")
except Exception as e:
    print(f"Validation failed: {e}")
```
